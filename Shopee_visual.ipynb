```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
```

```python
df = pd.read_csv('D:\\raw\\shopeep_koreantop_clothing_shop_data.csv')
df
```

```python
join_data = df.groupby('join_year')['shopid'].count()
join_data = pd.DataFrame(join_data )
join_data 
```

```python
plt.bar(join_data.index, join_data['shopid'])
plt.title('Số lượng shop gia nhập theo các năm', fontsize =14)
plt.xlabel('Năm', fontsize = 12)

plt.ylabel('Số lượng', fontsize = 12)
plt.show()
```

- tỉ lệ phản hồi với số lượt khách hàng đánh giá tốt

```python
plt.scatter(df['response_rate'], df['rating_good'])
plt.title('Mối tương quan giữa tỉ lệ phản hồi với số lượt khách hàng đánh giá tốt', fontsize = 15)

plt.xlabel('Tỉ lệ', fontsize = 12)
plt.ylabel('Số lượt', fontsize = 12)

fig = plt.gcf()
fig.set_size_inches(13,6.5)
plt.show()
```

- thời gian phản hồi (đơn vị giây) với số lượt khách hàng đánh giá xấu

```python
for i in df.index:
    hh, mm, ss = df['response_time'].iloc[i].split(':')
    df['response_time'].iloc[i] = int(hh) * 3600 + int(mm) * 60 + int(ss)
df
```

```python
plt.scatter(df['response_time'], df['rating_bad'])
plt.title('Mối tương quan giữa thời gian phản hồi (đơn vị giây) với số lượt khách hàng đánh giá xấu', fontsize = 15)

plt.xlabel('Giây', fontsize = 12)

plt.ylabel('Số lượt', fontsize = 12)

fig = plt.gcf()
fig.set_size_inches(13,6.5)
plt.show()
```
