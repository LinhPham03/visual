```python
df = pd.read_csv('D:\\raw\\shopeep_koreantop_clothing_shop_data.csv')
df
```

```python
join_data = df.groupby('join_year')['shopid'].count()
join_data = pd.DataFrame(join_data )
join_data 
```

- thá»i gian pháº£n há»“i (Ä‘Æ¡n vá»‹ giÃ¢y) vá»›i sá»‘ lÆ°á»£t khÃ¡ch hÃ ng Ä‘Ã¡nh giÃ¡ xáº¥u

```python
for i in df.index:
    hh, mm, ss = df['response_time'].iloc[i].split(':')
    df['response_time'].iloc[i] = int(hh) * 3600 + int(mm) * 60 + int(ss)
df
```

```python
plt.scatter(df['response_time'], df['rating_bad'])
plt.title('Má»‘i tÆ°Æ¡ng quan giá»¯a thá»i gian pháº£n há»“i (Ä‘Æ¡n vá»‹ giÃ¢y) vá»›i sá»‘ lÆ°á»£t khÃ¡ch hÃ ng Ä‘Ã¡nh giÃ¡ xáº¥u', fontsize = 15)

plt.xlabel('GiÃ¢y', fontsize = 12)

plt.ylabel('Sá»‘ lÆ°á»£t', fontsize = 12)

fig = plt.gcf()
fig.set_size_inches(13,6.5)
plt.show()
```
