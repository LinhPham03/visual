# 2. so sánh giá nhà trung bình trên 1 m2 giữa các hình thức nhà

```python
df = df.dropna(subset = ['area', 'price'])
df
```

```python
df['price/m2'] = round(df['price']/ df['area'],2)
df
```

```python
ngo = df[df['property_type'] == 'trong ngo']
ngo
```

```python
pho = df[df['property_type'] == 'mat pho']
pho
```

```python
inp = []
ngo_sum = 0
pho_sum = 0

for i in range(406):
    a = ngo['price/m2'].iloc[i]
    ngo_sum =+ a
for i in range(209):
    b = pho['price/m2'].iloc[i]
    pho_sum =+ b

inp.append(pho_sum)
inp.append(ngo_sum)
inp
```

```python
x = df['property_type'].unique()
y = inp
plt.bar(x, y)
plt.show()
```
