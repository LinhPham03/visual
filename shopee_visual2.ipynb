```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('D:\\raw\\shopeep_koreantop_clothing_shop_data.csv')
df
```

```python
join_data = df.groupby('join_year')['shopid'].count()
join_data = pd.DataFrame(join_data )
join_data 
```

# Câu 1

```python
fig, ax = plt.subplots(1,2)

ax[0].bar(join_data.index, join_data['shopid'],width = 0.5)
ax[0].set_ylabel('Số lượng')
ax[0].set_title('Số lượng shop gia nhập theo các năm', fontsize = 14)


ax[1].plot(join_data.index, join_data['shopid'])

fig.set_size_inches(18.5,6.5)
plt.show()
```

# Câu 2

```python
for i in df.index:
    hh, mm, ss = df['response_time'].iloc[i].split(':')
    df['response_time'].iloc[i] = int(hh) * 3600 + int(mm) * 60 + int(ss)
df
```

```python
fig, ax = plt.subplots(1,2)

ax[0].scatter(df['response_rate'], df['rating_good'])
ax[0].set_title('Tỷ lệ phản hồi vs Số lượt khách hàng đánh giá tốt', fontsize = 14)

ax[1].scatter(df['response_time'], df['rating_good'])
ax[1].set_title('Thời gian phản hồi vs Số lượt khách hàng đánh giá tốt', fontsize = 14)

fig.set_size_inches(18.5,6.5)
plt.show()
```
